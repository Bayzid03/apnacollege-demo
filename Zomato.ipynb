If we want to see which country has more rating, we need to use 'Aggregate rating','Rating color','Rating text' these columns.

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

#Observation
1.When rating are between 4.5 to 4.9 -------->>> Excellent
1.When rating are between 4.0 to 4.4 -------->>> Excellent
1.When rating are between 3.5 to 3.9 -------->>> Excellent
1.When rating are between 2.5 to 3.4 -------->>> Excellent
1.When rating are between 1.8 to 2.4 -------->>> Excellent

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue="Rating color",data=ratings, palette=['black','red','orange','yellow','green','blue'])
```
